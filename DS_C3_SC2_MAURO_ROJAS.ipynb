f) Determinar la cantidad de vacunas aplicadas el día 29/01/21 en todo el mundo.

```python
import pandas as pd
fecha_dada = '29/01/21'
vacunasDF = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/country_vaccinations.csv")
# Aseguramos que el campo 'date' sea del tipo datetime
vacunasDF['date'] = pd.to_datetime(vacunasDF['date'])
# Filtra el DataFrame para obtener las filas que coincidan con la fecha dada
filas_con_fecha_dada = vacunasDF[vacunasDF['date'] == fecha_dada]
# Suma los valores de la columna "daily_vaccinations" para las filas correspondientes a la fecha dada
suma_valores = filas_con_fecha_dada['daily_vaccinations'].sum()
# Imprime la suma de los valores
print("La cantidad de vacunas a nivel mundial para la fecha", fecha_dada, "es:", suma_valores)
```

j) Generar un dataframe denominado antes20 con todos los registros que se hayan realizado antes del 20 de diciembre de 2020.

```python
import pandas as pd
vacunasDF = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/country_vaccinations.csv")
antes20 = vacunasDF.copy()  # Copia el DataFrame original
# Asegurar que la columna 'date' sea del tipo datetime
vacunasDF['date'] = pd.to_datetime(vacunasDF['date'])
# Filtra las filas que cumplen con la condición (antes del 20 de diciembre de 2020)
antes20 = vacunasDF[vacunasDF['date'] < '20/12/2020']
# Contiene solo los registros realizados antes del 20 de diciembre de 2020
print('*** Nuevo Dataframe con registros realizados antes del 20/12/2020 ***')
print(antes20)
```

k) Obtener un dataframe denominado pfizer con todos los registros donde se haya utilizado la vacuna Pfizer.

```python
import pandas as pd
vacunasDF = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/country_vaccinations.csv")
pfizer = vacunasDF.copy()  # Copia el DataFrame original
# Filtra las filas que cumplen con la condición vaccine contiene pfizer
pfizer = vacunasDF[vacunasDF.vaccines.str.contains('Pfizer')]
# Contiene solo los registros deseados
print('*** Nuevo Dataframe con registros donde figure pfizer ***')
print(pfizer)
```

l) Almacenar los dataframes generados (conDiferencias, conCantidad, antes20 y pfizer) en un archivo de Excel denominado resultadosReto.xlsx, donde cada dataframe ocupe una hoja diferente. Se recomienda ver la documentación de pd.ExcelWriter.

```python
import pandas as pd
vacunasDF = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/country_vaccinations.csv")

# conDiferencias: Registros con columna de diferencias de datos
conDiferencias = vacunasDF.copy()  # Copia el DataFrame original
# Calcula la columna 'diferencias' como la diferencia entre 'daily_vaccinations' y 'daily_vaccinations_raw'
conDiferencias['diferencias'] = vacunasDF['daily_vaccinations'] - vacunasDF['daily_vaccinations_raw']

# conCantidad: Registros nueva columna con cantidad de vacunas utilizadas por dia
conCantidad = vacunasDF.copy()  # Copia el DataFrame original
# Nueva columna concatenando otras dos
conCantidad['canVac']= 'Cantidad vacunas usadas por día: ' + vacunasDF['daily_vaccinations'].astype(str) + ', Tipo: '  + vacunasDF['vaccines']

# antes20: Registros antes del 20/12/2020
antes20 = vacunasDF.copy()  # Copia el DataFrame original
# Asegurar que la columna 'date' sea del tipo datetime
vacunasDF['date'] = pd.to_datetime(vacunasDF['date'])
# Filtra las filas que cumplen con la condición (antes del 20 de diciembre de 2020)
antes20 = vacunasDF[vacunasDF['date'] < '20/12/2020']

# pfizer: Registros donde figure pfizer
pfizer = vacunasDF.copy()  # Copia el DataFrame original
pfizer = vacunasDF[vacunasDF.vaccines.str.contains('Pfizer')]

# Nombre del archivo de Excel
nombre_archivo = 'c:\resultadosReto.xlsx'
# Nombre de la hoja de Excel
nombre_hoja = 'pfizer'

# Guarda los DataFrames en el archivo de Excel en la hoja especificada
with pd.ExcelWriter("/content/drive/MyDrive/Colab Notebooks/Ejercicio I.xlsx") as writer:
    conDiferencias.to_excel(writer, sheet_name="conDiferencias")
    conCantidad.to_excel(writer, sheet_name="conCantidad")
    antes20.to_excel(writer, sheet_name="antes20")
    pfizer.to_excel(writer, sheet_name="pfizer")

#print("DataFrame guardado en el archivo de Excel:", nombre_archivo)
```
