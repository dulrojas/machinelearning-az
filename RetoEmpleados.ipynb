# **Reto: Deserción de empleados**

1.    Importa las librerías requeridas.

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import OrdinalEncoder
from sklearn.decomposition import PCA
```

2. Lee el archivo CSV llamado empleadosRETO.csv y coloca los datos en un frame de Pandas llamado EmpleadosAttrition.

```python
EmpleadosAttrition = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/RetoEmpleados/empleadosRETO.csv")
EmpleadosAttrition
```

3. Elimina las columnas que, con alta probabilidad (estimada por ti), no tienen relación alguna con la salida. Hay algunas columnas que contienen información que no ayuda a definir el desgaste de un empleado, tal es caso de las siguientes:

  a.     EmployeeCount: número de empleados, todos tienen un 1

  b.     EmployeeNumber: ID del empleado, el cual es único para cada empleado

  c.     Over18: mayores de edad, todos dicen “Y”

  d.     StandardHours: horas de trabajo, todos tienen “80”

```python
EmpleadosAttrition = EmpleadosAttrition.drop(['EmployeeCount','EmployeeNumber','Over18','StandardHours'],axis=1)
EmpleadosAttrition
```

4. Analizando la información proporcionada, detectaste que no se cuenta con los años que el empelado lleva en la compañía y parece ser un buen dato. Dicha cantidad se puede calcular con la fecha de contratación ‘HiringDate’:

  a.     Crea una columna llamada Year y obtén el año de contratación del empleado a partir de su fecha ‘HiringDate’. No se te olvide que debe ser un entero.

  b.     Crea una columna llamada YearsAtCompany que contenga los años que el empleado lleva en la compañía hasta el año 2018. Para su cálculo, usa la variable Year que acabas de crear.

```python
# Identificar fechas inválidas
invalid_dates = EmpleadosAttrition['HiringDate'][~pd.to_datetime(EmpleadosAttrition['HiringDate'], errors='coerce').notnull()]
# Borrar registros con fechas inválidas
EmpleadosAttrition.drop(invalid_dates.index, inplace=True)
# Decisión tomada ya que solo existe un dato malo
```

```python
# Convertir la columna 'fecha_nacimiento' a tipo datetime
EmpleadosAttrition['HiringDate'] = pd.to_datetime(EmpleadosAttrition['HiringDate'], format='%m/%d/%Y')
# a. Obtener el año de contratación y almacenarlo en una nueva columna 'year'
EmpleadosAttrition['Year'] = EmpleadosAttrition['HiringDate'].dt.year.astype(int)
# b. Obtener los años de la persona hasta fin de 2018 y almacenarlos en una nueva columna 'edad_hasta_2018'
EmpleadosAttrition['YearsAtCompany'] = 2018 - EmpleadosAttrition['HiringDate'].dt.year
EmpleadosAttrition
```

5.    La DistanceFromHome está dada en kilómetros, pero tiene las letras “km” al final y así no puede ser entera:

  a.     Renombra la variable DistanceFromHome a DistanceFromHome_km.
  
  b.     Crea una nueva variable DistanceFromHome que sea entera, es decir, solo con números.

```python
# a. Renombrar la variable DistanceFromHome a DistanceFromHome_km
EmpleadosAttrition.rename(columns={'DistanceFromHome': 'DistanceFromHome_km'}, inplace=True)

# b. Crear una nueva variable DistanceFromHome que sea entera, es decir, solo con números
EmpleadosAttrition['DistanceFromHome'] = EmpleadosAttrition['DistanceFromHome_km'].str.replace('km', '').astype(int)
EmpleadosAttrition
```

6.    Borra las columnas Year, HiringDate y DistanceFromHome_km debido a que ya no son útiles.

```python
EmpleadosAttrition = EmpleadosAttrition.drop(['Year','HiringDate','DistanceFromHome_km'],axis=1)
EmpleadosAttrition
```

7.    Aprovechando los ajustes que se están haciendo, la empresa desea saber si todos los departamentos tienen un ingreso promedio similar. Genera una nuevo frame llamado SueldoPromedioDepto que contenga el MonthlyIncome promedio por departamento de los empleados y colócalo en una variable llamada SueldoPromedio. Esta tabla solo es informativa, no la vas a utilizar en el set de datos que estás construyendo.

```python
#SueldoPromedioDepto que contenga el MonthlyIncome promedio por departamento de los empleados y colócalo en una variable llamada SueldoPromedio.
# Calcula el sueldo promedio por departamento
SueldoPromedioDepto = EmpleadosAttrition.groupby('Department')['MonthlyIncome'].mean().reset_index()
SueldoPromedioDepto.columns = ['Department', 'SueldoPromedio']

print(SueldoPromedioDepto)
```

8.    La variable MonthlyIncome tiene un valor numérico muy grande comparada con las otras variables. Escala dicha variable para que tenga un valor entre 0 y 1.

```python
# Calcula el valor mínimo y máximo de MonthlyIncome
min_income = EmpleadosAttrition['MonthlyIncome'].min()
max_income = EmpleadosAttrition['MonthlyIncome'].max()

# Escala la variable MonthlyIncome entre 0 y 1
EmpleadosAttrition['MonthlyIncome'] = (EmpleadosAttrition['MonthlyIncome'] - min_income) / (max_income - min_income)

EmpleadosAttrition
```

9.    Todo parece indicar que las variables categóricas que quedan sí son importantes para obtener la variable de salida. Convierte todas las variables categóricas que quedan a numéricas:

  a.     BusinessTravel

  b.     Department

  c.     EducationField

  d.     Gender

  e.     JobRole

  f.      MaritalStatus

  g.     Attrition

```python
# a. BusinessTravel
# Remplace los datos vacíos
EmpleadosAttrition['BusinessTravel'].fillna('Unknown', inplace=True)
# Inicializa el codificador ordinal
ordinal_encoder = OrdinalEncoder(categories=[['Unknown','Non-Travel', 'Travel_Rarely', 'Travel_Frequently']])
# Aplica el ordinal encoding a la columna 'BusinessTravel'
EmpleadosAttrition['BusinessTravel'] = ordinal_encoder.fit_transform(EmpleadosAttrition[['BusinessTravel']])

#b. Department
# Aplicar one-hot encoding a la columna 'Department'
df_encoded = pd.get_dummies(EmpleadosAttrition['Department'], prefix='Department')
# Concatenar el DataFrame original con las nuevas columnas codificadas
EmpleadosAttrition = pd.concat([EmpleadosAttrition, df_encoded], axis=1)

# c. EducationField
# Aplicar one-hot encoding a la columna 'EducationField'
df_encoded = pd.get_dummies(EmpleadosAttrition['EducationField'], prefix='EducationField')
# Concatenar el DataFrame original con las nuevas columnas codificadas
EmpleadosAttrition = pd.concat([EmpleadosAttrition, df_encoded], axis=1)

# d.     Gender
# Aplicar one-hot encoding a la columna 'Gender'
df_encoded = pd.get_dummies(EmpleadosAttrition['Gender'], prefix='Gender')
# Concatenar el DataFrame original con las nuevas columnas codificadas
EmpleadosAttrition = pd.concat([EmpleadosAttrition, df_encoded], axis=1)

# e.     JobRole
# Aplicar one-hot encoding a la columna 'JobRole'
df_encoded = pd.get_dummies(EmpleadosAttrition['JobRole'], prefix='JobRole')
# Concatenar el DataFrame original con las nuevas columnas codificadas
EmpleadosAttrition = pd.concat([EmpleadosAttrition, df_encoded], axis=1)

# f.      MaritalStatus
# Remplace los datos vacíos
EmpleadosAttrition['MaritalStatus'].fillna('Unknown', inplace=True)
# Aplicar one-hot encoding a la columna 'JobRole'
df_encoded = pd.get_dummies(EmpleadosAttrition['MaritalStatus'], prefix='MaritalStatus')
# Concatenar el DataFrame original con las nuevas columnas codificadas
EmpleadosAttrition = pd.concat([EmpleadosAttrition, df_encoded], axis=1)

# g.     Attrition
# Inicializa el codificador ordinal
ordinal_encoder = OrdinalEncoder(categories=[['No','Yes']])
# Aplica el ordinal encoding a la columna 'BusinessTravel'
EmpleadosAttrition['Attrition'] = ordinal_encoder.fit_transform(EmpleadosAttrition[['Attrition']])

EmpleadosAttrition
```

12.  Crea una nueva variable llamada EmpleadosAttritionPCA formada por los componentes principales del frame EmpleadosAttritionFinal. Recuerda que el resultado del proceso PCA es un numpy array, por lo que, para hacer referencia a una columna, por ejemplo, la 0, puedes usar la instrucción EmpleadosAttritionPCA[:,0]).

```python
pca = PCA()
pca.fit(EmpleadosAttritionFinal)
EmpleadosAttritionPCA = pca.components_
EmpleadosAttritionPCA
```

13.  Agrega el mínimo número de Componentes Principales en columnas del frame EmpleadosAttritionPCA que logren explicar el 80% de la varianza, al frame EmpleadosAttritionFinal. Puedes usar la instrucción assign, columna por columna, llamando a cada una C0, C1, etc., hasta las que vayas a agregar.

```python
pca = PCA()
pca.fit(EmpleadosAttritionFinal)
print(pca.explained_variance_)
```

```python
EmpleadosAttritionPCA = pca.transform(EmpleadosAttritionFinal)
print(EmpleadosAttritionPCA)
```

```python
# Lista de nombres de columna
EmpleadosAttritionFinal = EmpleadosAttritionFinal.assign(C0 = EmpleadosAttritionPCA[:, 0])
EmpleadosAttritionFinal = EmpleadosAttritionFinal.assign(C1 = EmpleadosAttritionPCA[:, 1])
EmpleadosAttritionFinal = EmpleadosAttritionFinal.assign(C2 = EmpleadosAttritionPCA[:, 2])
EmpleadosAttritionFinal = EmpleadosAttritionFinal.assign(C3 = EmpleadosAttritionPCA[:, 3])
EmpleadosAttritionFinal = EmpleadosAttritionFinal.assign(C4 = EmpleadosAttritionPCA[:, 4])

EmpleadosAttritionFinal
```
